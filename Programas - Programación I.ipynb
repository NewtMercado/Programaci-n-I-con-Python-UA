```python
import numpy as np

def ProductoInterno(A, B, n):
    Suma = 0
    for i in range(n):
        Suma += A[i] * B[i]
    return Suma

def ProductoExterno(A, B):
    m = len(A)
    n = len(B)
    C = np.zeros((m, n))  # Creamos una matriz de ceros para almacenar el resultado
    for i in range(m):
        for j in range(n):
            C[i, j] = ProductoInterno(A, B, n)
    return C

# Ejemplo de uso
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
resultado = ProductoExterno(A, B)
print(resultado)
```
