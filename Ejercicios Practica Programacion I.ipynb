```python
while True:
    Numero = float(input("Ingrese un numero real entre 0 y 1: "))
    if 0 <= Numero <= 1:
        print("Ha elegido un buen número :)")
        break
    else:
        print("Por favor, ingrese un número válido entre 0 y 1.")

ArcotanDeNumero = 0
Tol = 0.0001
n = 0
while True:
    termo = (-1) ** n * (Numero ** (2 * n + 1) / (2 * n + 1))
    if abs(termo) < Tol:
        break
    ArcotanDeNumero += termo
    n += 1

print("La aproximación de arcotangente de", Numero, "es:", ArcotanDeNumero)
```
