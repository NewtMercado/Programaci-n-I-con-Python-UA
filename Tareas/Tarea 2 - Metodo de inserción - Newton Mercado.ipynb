## Metodo para lista

```python
# Metodo de Ordenacion:
def MetodoInsercion(A):
  key= " "
  j=2
  for j in range(len(A)):
    key= A[j]
    i=j-1
    while i > 0 and A[i]>key:
      A[i+1]=A[i]
      i=i-1
    A[i+1]=key
  return A
```

```python
A= [5, 2, 4, 6, 1, 3]
MetodoInsercion(A)
Numero = A[0]
del A[0]
A.append(Numero)
MetodoInsercion(A)
```

## Metodo para array

```python
# Metodo de Ordenacion:
def MetodoOrdenacion(A):
  Longitud= len(A)
  for i in range(1, Longitud):
    ValorInsertado = A[i]
    ValorIndiceInsertar = i

    while ValorIndiceInsertar > 0 and A[ValorIndiceInsertar - 1] > ValorInsertado:
      # Se define el espacio para hacer la inserci√≥n:
      A[ValorIndiceInsertar] = A[ValorIndiceInsertar - 1]
      # Se actualiza la variable:
      ValorIndiceInsertar = ValorIndiceInsertar-1
      # Inserci√≥n del el valor en el indice:
      A[ValorIndiceInsertar] = ValorInsertado
  return A
```

```python
A= [2, 4, 1, 3, 6]
MetodoOrdenacion(A)
```
