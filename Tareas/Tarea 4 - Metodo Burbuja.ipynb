```python
def MetodoBurbuja(A):
  i=1
  j= len(A)
  for i in range(len(A)-1):
    for j in range(len(A)-1):
      if A[j] > A[j+1]:
        A[j], A[j+1] = A[j+1], A[j]
  return A
```

```python
A= [2, 4, 1, 3, 6]
MetodoBurbuja(A)
```
