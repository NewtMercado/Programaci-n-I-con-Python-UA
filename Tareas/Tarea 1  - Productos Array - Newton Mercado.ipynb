```python
# Función para calcular el producto interno entre dos vectores
def producto_interno(A, B):
    if len(A) == len(B):
        return np.dot(A, B) #funcion encotrada en Numpy
    else:
        return "El número de elementos de los vectores deben ser iguales"
```

```python
# Función para realizar la multiplicación de una matriz por un vector
def matriz_por_vector(A, B):
    if len(A[0]) != len(B):
        return "El número de columnas de la matriz debe ser igual a la longitud del vector."

    m= len(A)
    C = np.zeros(m)

    for i in range(m):
        C[i] = producto_interno(A[i], B)
    return C
```

```python
# Función para realizar la multiplicación de una matriz por matriz
def matriz_por_matriz(A, B):
    m = len(A)
    n1 = len(A[0])
    n2 = len(B)
    p = len(B[0])

    if n1 != n2:
        return "El número de columnas de la primera matriz debe ser igual al número de filas de la segunda matriz."

    C = np.zeros((m, p))

    for i in range(p):
        C = matriz_por_vector(A, B[:, i])

    return C
```

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([[1, 2, 3], [4, 5, 6]])
resultado = matriz_por_matriz(A, B)
print(resultado)
```

```python
# Definir la matriz y el vector
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([1, 2])

resultado = matriz_por_vector(A, B)
print(resultado)
```
