```python
# SUMA DE POLINOMIOS FORMA 1:
def SumaPoly_0(P, Q):
    SumaPoly = []
    for i, k in zip(P, Q):
        SumaPoly.append(i + k)
    return SumaPoly #esta sumando los polinomios de atras hacia delante
```

```python
# SUMA DE POLINOMIOS FORMA 2:
def SumaPoly(P, Q):
  len_P = len(P)
  len_Q = len(Q)
  max_len = max(len_P, len_Q) #Grado
  SumaPoly = []

  for i in range(max_len):
    coeficiente_P = P[i] if i < len_P else 0 # los condicionales me ayudan a rellenar la lista con ceros en caso de ser de menor grado uno de los polinomios
    coeficiente_Q = Q[i] if i < len_Q else 0 # los condicionales me ayudan a rellenar la lista con ceros en caso de ser de menor grado uno de los polinomios
    suma_coeficientes = coeficiente_P + coeficiente_Q
    SumaPoly.append(suma_coeficientes)

  return SumaPoly #esta sumando los polinomios de atras hacia delante
```

```python
# PRODUCTOS DE POLINOMIOS:
def ProductoPolinomios(P, Q):
    gradoP = len(P) - 1
    gradoQ = len(Q) - 1
    gradoResultante = gradoP + gradoQ
    producto = [0] * (gradoResultante + 1)
    for i in range(gradoP + 1):
        for j in range(gradoQ + 1):
            producto[i + j] += P[i] * Q[j]
    return producto #esta haciendo el producto de los polinomios de atras hacia delante
```

```python
def EvaluarPolinomioHorner(coeficientes, x):
    resultado = 0
    for coeficiente in coeficientes:
        resultado = resultado * x + coeficiente
    return resultado
```

```python
# Parte principal del programa

# Ingreso del polinomio P
print("Ingrese los coeficientes del polinomio P:")
GradoPoly_P = int(input("Ingrese el grado del polinomio P: "))
P = []
for i in range(GradoPoly_P + 1):
    p_i = float(input(f"Ingrese el coeficiente del término de grado {i}: "))
    P.append(p_i)
print("P:", P)

# Ingreso del polinomio Q
print("\nIngrese los coeficientes del polinomio Q:")
GradoPoly_Q = int(input("Ingrese el grado del polinomio Q: "))
Q = []
for k in range(GradoPoly_Q + 1):
    q_i = float(input(f"Ingrese el coeficiente del término de grado {k}: "))
    Q.append(q_i)
print("Q:", Q)

# Suma de polinomios
resultado_suma = SumaPoly(P, Q)
print("\nLa suma de los polinomios es:", resultado_suma)

# Producto de polinomios
resultado_producto = ProductoPolinomios(P, Q)
print("\nEl producto de los polinomios es:", resultado_producto)

# Evaluación de un polinomio mediante el algoritmo de Horner
x = float(input("\nIngrese el valor de x para evaluar el polinomio: "))
resultado_evaluacion = EvaluarPolinomioHorner(P, x)
print(f"El polinomio evaluado en x={x} es: {resultado_evaluacion}")
```
