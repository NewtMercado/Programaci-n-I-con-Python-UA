```python
class Carro:
  def __init__ (self,Marca, Color, Motor):
    self.Marca= Marca
    self.Color= Color
    self.Motor= Motor

class CarroDeportivo (Carro):
  def __init__ (self,Marca,Color,Motor,Abs,Automatico):
    super().__init__(Marca,Color,Motor)
    self.Abs= Abs
    self.Automatico= Automatico

Carro1= CarroDeportivo("Volovo","Sepia","V9",True,1)

print("El Carro 1 es:", Carro1.Marca,Carro1.Color,"con",Carro1.Abs)
```
